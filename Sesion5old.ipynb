## Exploration of network communities

* **Average clustering coefficient**: is the average of the local clustering coefficients of the nodes. If all the neighbors of every node are connected among one another, you get 1; if none of them are connected you get zero.

```python
# count_zeros=False -> to make results compatible with R!
nx.average_clustering(topsNet_giant,count_zeros=False)
```

* **Transitivity**: Tendency of local clustering.

```python
# Probability that two tech guys with a common colleague, are also colleagues.
nx.transitivity(topsNet_giant)
```

* **Assortativity (degree)**: it is a measure to see if nodes are connecting to other nodes similar in degree.  Closer to 1 means higher assortativity, closer to -1 diassortativity; while 0 is no assortitivity.

```python
nx.degree_assortativity_coefficient(topsNet_giant)
```

You can also compute assortativity using an attribute of interest:

```python
nx.attribute_assortativity_coefficient(topsNet_giant,'male') # positive if tend to connect with same sex.
```

```python
nx.attribute_assortativity_coefficient(topsNet_giant,'popularity') # positive if tend to connect with same sex.
```

A **clique** can be understood a community of nodes where all of them are connected to one another.

* How many cliques do we have?

```python
# len(list(nx.enumerate_all_cliques(topsNet)))
len(list(nx.find_cliques_recursive(topsNet_giant)))
```

If a clique in the network can not be bigger is you add another node, then you have a **maximal clique**.

<center><img src="https://github.com/EvansDataScience/CTforGA_Networks/raw/main/cliqueMaximal.png" width="500"></center>

* How many maximal cliques are there in this network?

```python
# sum(1 for _ in nx.find_cliques(topsNet))
```

This is how you see every maximal clique:

```python
for a in nx.find_cliques_recursive(topsNet_giant):
    print (a)
```

You can find the size of the _maximum clique_, largest clique in the graph, like this:

```python
maxsize_clique=max(len(c) for c in nx.find_cliques_recursive(topsNet_giant))
maxsize_clique
```

You can see each maximum clique like this:

```python
[c for c in nx.find_cliques_recursive(topsNet_giant) if len(c) == maxsize_clique]
```

If a network presents cliques makes you suspect that there can be **communities**.

This is a huge field of research, let me me show you one of the algorithms known as the Louvain method.



```python
nx.draw_kamada_kawai(topsNet_giant)
```

```python
# Calculating the betweenness centrality
btw_centrality = nx.algorithms.centrality.edge_betweenness_centrality(topsNet_giant)
list(btw_centrality.items())[:4]
```

```python


# Sorting based on the betweenness centraliy and displaying the first 10 edges.
sorted(btw_centrality.items(), key = lambda item:item[1], reverse = True)[0:10]
```

```python
bye_edge_1 = sorted(btw_centrality.items(), key = lambda item:item[1], reverse = True)[0]
bye_edge_1
```

```python
pos=nx.kamada_kawai_layout(topsNet_giant)
nx.draw(topsNet_giant,pos=pos, width=0.1,edge_color='grey')
nx.draw_networkx_edges(topsNet_giant,pos=pos,edgelist=[bye_edge_1[0]],width=6,edge_color='red' );
```

```python
topsNet_giant_1=topsNet_giant.copy()
topsNet_giant_1.remove_edge(*bye_edge_1[0])
btw_centrality = nx.algorithms.centrality.edge_betweenness_centrality(topsNet_giant_1)
bye_edge_2=sorted(btw_centrality.items(), key = lambda item:item[1], reverse = True)[0]

nx.draw(topsNet_giant,pos=pos, width=0.1,edge_color='grey')
nx.draw_networkx_edges(topsNet_giant,pos=pos,edgelist=[bye_edge_1[0],bye_edge_2[0]],width=6,edge_color='red' );
```

```python
topsNet_giant_2=topsNet_giant_1.copy()
topsNet_giant_2.remove_edge(*bye_edge_2[0])
btw_centrality = nx.algorithms.centrality.edge_betweenness_centrality(topsNet_giant_2)
bye_edge_3=sorted(btw_centrality.items(), key = lambda item:item[1], reverse = True)[0]

nx.draw(topsNet_giant,pos=pos, width=0.1,edge_color='grey')
nx.draw_networkx_edges(topsNet_giant,pos=pos,edgelist=[bye_edge_1[0],bye_edge_2[0],bye_edge_3[0]],width=6,edge_color='red' );
```

```python
topsGirvNew = nx.algorithms.community.centrality.girvan_newman(topsNet_giant)
topsGirvNew_result=[x for x in topsGirvNew if len(x)==4][0]
```

```python
girvanNewman_attr={z:x for x,y  in enumerate(topsGirvNew_result) for z in y }
girvanNewman_attr
```

```python
nx.set_node_attributes(topsNet_giant, girvanNewman_attr,'girvanNewman')
```

```python
list(dict(topsNet_giant.nodes(data=True)).items())[:4]
```

```python
pos=nx.spring_layout(topsNet_giant,k=3) 
plt.figure(figsize=(8,8))
color_map = plt.get_cmap("cool")
valuesForColors=[n[1]['girvanNewman'] for n in topsNet_giant.nodes(data=True)]
nx.draw(topsNet_giant,pos=pos,node_color=valuesForColors,cmap=color_map,with_labels=True,edge_color='lightblue')
```

```python
topsLouvain=nx.community.louvain_communities(topsNet_giant, seed=123)
louvain_attr={z:x for x,y  in enumerate(topsLouvain) for z in y }
louvain_attr
```

```python
topsLouvain=nx.community.louvain_communities(topsNet_giant, seed=123)
louvain_attr={z:x for x,y  in enumerate(topsLouvain) for z in y }
nx.set_node_attributes(topsNet_giant, louvain_attr,'louvain')

pos=nx.spring_layout(topsNet_giant,k=3) 
plt.figure(figsize=(8,8))
color_map = plt.get_cmap("cool")
valuesForColors=[n[1]['louvain'] for n in topsNet_giant.nodes(data=True)]
nx.draw(topsNet_giant,pos=pos,node_color=valuesForColors,cmap=color_map,with_labels=True,edge_color='lightblue')
```

```python
communities = list(nx.community.girvan_newman(topsNet_giant))
modularity_df = pd.DataFrame(
    [
        [k + 1, nx.community.modularity(topsNet_giant, communities[k])]
        for k in range(len(communities))
    ],
    columns=["k", "modularity"],
)


# Plot change in modularity as the important edges are removed
modularity_df.plot.bar(
    x="k",
    color="#F2D140",
    title="Modularity Trend for Girvan-Newman Community Detection",
)
```
