# Síntesis y Aplicación

Los grafos son aplicables a todo campo donde haya relación entre actores, agentes, nodos, etc. En el campo de la innovación, hay característisticas particulares que pueden ser resumidas así:


### Grupo de Inventores (I): Los Orígenes de la Novedad

Los **Inventores** son el corazón de la innovación. Su importancia radica en su rol como **generadores de conocimiento original y disruptivo**.

* **Fuentes de Conocimiento:** Son quienes crean nuevas ideas, patentes, publicaciones científicas y prototipos. Sin ellos, el flujo de "novedad" simplemente no existiría.
* **Validación Técnica:** El prestigio de un inventor (alto índice Scopus, patentes exitosas, premios) funciona como un **sello de calidad y viabilidad técnica**. Otros actores confían en su experiencia para decidir si una idea merece ser explorada o financiada.
* **Atracción de Colaboración:** Los inventores con alto prestigio atraen a otros investigadores, emprendedores y empresas, buscando colaborar y acceder a su conocimiento de vanguardia. Si un inventor de alto prestigio está aislado o sobrecargado, el flujo de nuevas ideas puede estancarse.

---

### Grupo de Emprendedores (E): Los Catalizadores de la Comercialización

Los **Emprendedores** son los que transforman las ideas en valor. Son el puente esencial entre el laboratorio o la mente del inventor y el mercado.

* **Agentes de Transformación:** Toman los inventos, les dan forma de producto o servicio, y desarrollan modelos de negocio. Sin ellos, muchas innovaciones se quedarían solo en papel.
* **Atracción de Capital y Talento:** El prestigio de un emprendedor (experiencia previa exitosa, capacidad de liderazgo) es clave para **atraer financiación y talento** para escalar la innovación.
* **Conexión con el Mercado:** Son los principales impulsores del **flujo de la innovación hacia los usuarios y clientes**, detectando necesidades y validando soluciones. Si los emprendedores carecen de prestigio o de conexiones, las innovaciones pueden quedarse "atascadas" en etapas tempranas.

---

### Grupo de Consultores (C): Los Facilitadores y Conectores Estratégicos

Los **Consultores** son los engranajes que lubrican y conectan la maquinaria de la innovación, proporcionando experticia y red.

* **Asesoramiento Experto:** Ofrecen conocimiento especializado en áreas clave como estrategia, propiedad intelectual, marketing o gestión, crucial para superar obstáculos.
* **Puentes de Conocimiento:** Un consultor con alto prestigio como "C_Julia" no solo aporta conocimiento, sino también una **vasta red de contactos**. Pueden conectar inventores con financistas, emprendedores con expertos técnicos, facilitando el **flujo de información y oportunidades** a través de barreras sectoriales o geográficas.
* **Reducción de Incertidumbre:** Su experiencia y reputación ayudan a validar decisiones y reducir la incertidumbre en un proceso inherentemente riesgoso como la innovación. Si estos nodos son escasos o inaccesibles, el flujo puede ralentizarse por falta de dirección o conexiones.

---

### Grupo de Financistas (F): Los Impulsores de Recursos

Los **Financistas** son el motor económico de la innovación, proveyendo el capital necesario para que las ideas despeguen y crezcan.

* **Aportación de Capital:** Su función principal es inyectar el **flujo de recursos financieros** que permite investigar, desarrollar prototipos, comercializar productos y escalar operaciones.
* **Validación de Oportunidades:** El prestigio de un financista (historial de inversiones exitosas) sirve como una **señal de validación** para el mercado y otros inversores. Su compromiso con un proyecto atrae a más capital y confianza.
* **Acceso a Redes de Negocio:** A menudo, los financistas de prestigio no solo aportan dinero, sino también mentoría, experiencia de negocio y acceso a redes estratégicas, que son un **flujo de conocimiento y conexiones** vital. Un cuello de botella en este grupo puede ahogar la innovación antes de que vea la luz.

---

### Grupo de Administradores (A): Los Habilitadores del Entorno

Los **Administradores** (ya sean de instituciones académicas, gubernamentales o grandes corporaciones) crean y mantienen el entorno en el que la innovación puede florecer o marchitarse.

* **Establecimiento de Políticas y Recursos:** Definen las reglas, asignan presupuestos, gestionan la infraestructura y establecen los incentivos que permiten la investigación, el desarrollo y la colaboración.
* **Gestión del Ecosistema:** Su prestigio se relaciona con su capacidad para fomentar un ambiente propicio para la innovación, eliminando barreras burocráticas y promoviendo la colaboración. Son facilitadores del **flujo de permisos, recursos institucionales y marcos de trabajo**.
* **Visión Estratégica:** A menudo, son los responsables de la visión a largo plazo y de la asignación estratégica de recursos. Si la administración carece de prestigio o visión, puede crear un cuello de botella por la falta de un entorno favorable o el acceso limitado a recursos internos.

---

### Grupo de Usuarios (U): Los Validadores y Adaptadores Finales

Aunque a menudo se les ve al final del ciclo, los **Usuarios** son cruciales para el **flujo de retroalimentación** y la **adopción de la innovación**.

* **Validación del Mercado:** Son quienes finalmente adoptan o rechazan una innovación, proporcionando la retroalimentación esencial que valida si una idea realmente resuelve un problema. Su "prestigio" puede medirse por su capacidad de ser "lead users" o influenciadores en sus comunidades.
* **Fuente de Nuevas Ideas:** Los usuarios avanzados o con alto "prestigio de influencia" pueden ser una fuente invaluable de nuevas ideas y necesidades no satisfechas, impulsando el **flujo de futuros desarrollos innovadores**.
* **Difusión Final:** La adopción y recomendación por parte de los usuarios (especialmente los influyentes) es el flujo final que lleva la innovación a una escala masiva. Si la innovación no llega a este grupo o no es bien recibida, todo el esfuerzo anterior puede ser en vano.

Creemos una función para describir esta red:

````python
# Métricas
def analizar_red(G, nombre):
    """Retorna un diccionario con todas las métricas calculadas."""
    largest_component_nodes = max(nx.connected_components(G), key=len)
    G_cgc = G.subgraph(largest_component_nodes).copy()
    metricas = {
        'nombre': nombre,
        'nodos_total': len(G),
        'componentes': nx.number_connected_components(G),
        'nodos_gigante': len(max(nx.connected_components(G), key=len)),
        'grado_promedio_gigante': sum(dict(G_cgc.degree()).values()) / len(G_cgc),
        'clustering_gigante': nx.average_clustering(G_cgc),
        'distancia_promedio_gigante': nx.average_shortest_path_length(G_cgc)
    }
    metricas['porcentaje_gigante'] = (metricas['nodos_gigante'] / metricas['nodos_total']) * 100
    return metricas
````

Usemosla:

### ¿Por qué estas métricas básicas son importantes?

#### 1. Nodos Totales (`nodos_total`)
* **Importancia:** Saber el número total de nodos nos da una idea de la **escala** del ecosistema de innovación. Esta cifra es el punto de partida para todas las demás proporciones y densidades. Un número mayor de nodos puede indicar un ecosistema de innovación más diverso y potencialmente más rico en ideas, pero también más complejo de gestionar.

#### 2. Componentes Conectados (`componentes_total`)
* **Importancia:** Nos dice si la red está fragmentada o si forma un todo cohesionado. Varios componentes indican que hay **"islas" de actores que no están conectados entre sí**. En una red de innovación, esto es un gran problema, ya que el conocimiento y las oportunidades no pueden fluir entre estas islas. Si una idea innovadora surge en una isla, es posible que nunca llegue a los financistas o emprendedores en otra. Un alto número de componentes indica barreras estructurales en una sociedad al flujo de innovación.

#### 3. Nodos en el Componente Conectado Más Grande (CGC) (`nodos_en_cgc` y `porcentaje_cgc`)
* **Importancia:** Este es el **núcleo funcional**. Incluso si se tienen muchos componentes, el CGC representa la porción más grande y robusta donde el **flujo es realmente posible**.
    * `nodos_en_cgc`: Dice cuántos actores están efectivamente "en el juego" de la innovación interconectada.
    * `porcentaje_cgc`: Da una idea de la **salud general de la red**. Un porcentaje bajo significa que una gran parte de tus actores están aislados del flujo principal. Un porcentaje alto indica que la mayoría (o todos) los actores pueden, en teoría, interactuar entre sí.

#### 4. Grado Promedio (del CGC) (`grado_promedio_cgc`)
* **Importancia:** El grado de un nodo es simplemente el número de conexiones que tiene. El grado promedio te dice la **conectividad general de los actores en el núcleo de tu red**.
    * Un grado promedio más alto indica una red más **densa y rica en conexiones**, lo que facilita el intercambio de información y el acceso a recursos.
    * En una red de innovación, un grado promedio bajo podría significar que el conocimiento se mueve lentamente o que los actores tienen dificultades para encontrar colaboradores o recursos. 

#### 5. Coeficiente de Clustering (del CGC) (`clustering_cgc`)
* **Importancia:** Esta métrica mide la **tendencia de los nodos a formar grupos o "cliques"**. Responde a la pregunta: "Los amigos de mis amigos, ¿también son mis amigos?".
    * Un **alto coeficiente de clustering** (cercano a 1) sugiere que la red está formada por comunidades o subgrupos muy unidos. Esto puede fomentar la **confianza**, el **intercambio de conocimiento tácito** y la **validación rápida** dentro de esos grupos. Sin embargo, un *clustering* excesivo puede crear **"silencios" o barreras al flujo de nuevas ideas entre comunidades diferentes** (cuello de botella por "endogamia").
    * Un **bajo coeficiente de clustering** (cercano a 0) sugiere una red más "aleatoria" o menos estructurada en grupos. Esto puede favorecer la **difusión de la novedad** a través de la red, pero podría carecer de la confianza y el soporte necesarios para proyectos de innovación más complejos.

#### 6. Distancia Promedio del Camino Más Corto (del CGC) (`distancia_promedio_cgc`)
* **Importancia:** Este es un proxy de **eficiencia de la comunicación y el flujo de información** entre cualquier par de nodos en el núcleo de tu red. Es el número promedio de "saltos" o pasos que se necesitan para ir de un nodo a otro.
    * Una **distancia promedio corta** (como 2.279) indica que la información, las ideas y los recursos pueden **moverse muy rápidamente** a través de la red. Esto es altamente deseable en una red de innovación, ya que acelera la colaboración, la retroalimentación y el desarrollo de nuevos productos o servicios.
    * Una **distancia promedio larga** indicaría un cuello de botella por **ineficiencia en la comunicación**, donde las ideas tardan mucho en llegar a quienes las necesitan o pueden actuar sobre ellas.

En resumen, una red con muchos componentes (problema estructural), un grado promedio bajo (pocas conexiones), o una distancia de camino largo (flujo lento) te indicaría dónde enfocar tus esfuerzos para mejorar la conectividad y la eficiencia de la innovación.

````python
import matplotlib.pyplot as plt
import numpy as np 

print(f"Red creada con {innovation_network.number_of_nodes()} nodos y {innovation_network.number_of_edges()} aristas.")

# --- Configuración para la Visualización ---

# Asignar colores basados en roles para la visualización
node_colors = {
    'Inventor': 'skyblue',
    'Emprendedor': 'orange',
    'Consultor': 'lightgreen',
    'Financista': 'red',
    'Administrador': 'purple',
    'Usuario': 'gray'}

# Crear una lista de colores de nodos en el mismo orden que los nodos del grafo
colors = [node_colors[innovation_network.nodes[node]['role']] for node in innovation_network.nodes()]

# Obtener los valores de prestigio de todos los nodos
prestigio_values = [innovation_network.nodes[node]['prestigio'] for node in innovation_network.nodes()]


# --- Escalar el tamaño del nodo según el prestigio ---
# Los valores de prestigio pueden ser muy variados, por lo que es buena idea escalarlos
# Para una escala de prestigio de 0 a 100, un factor de 5-10 es un buen punto de partida.
# Vamos a normalizar primero el prestigio y luego escalarlo.
min_prestigio = min(prestigio_values)
max_prestigio = max(prestigio_values)

# Normalizar entre 100 a 1800
scaled_prestigio = [
    ((p - min_prestigio) / (max_prestigio - min_prestigio) * (1800 - 100) + 100)
    for p in prestigio_values
]


# Configurar las etiquetas de los nodos (mostrar el nombre del nodo)
labels = {node: node for node in innovation_network.nodes()}

# --- Generar el Gráfico ---
plt.figure(figsize=(14, 12)) # Aumentar un poco el tamaño de la figura
pos = nx.spring_layout(innovation_network, seed=50) # Layout para visualizar (seed para reproducibilidad)

# Dibuja los nodos con el tamaño escalado por prestigio
nx.draw_networkx_nodes(innovation_network, pos, node_color=colors, node_size=scaled_prestigio, alpha=0.9)
nx.draw_networkx_edges(innovation_network, pos, edge_color='gray', alpha=0.6)
nx.draw_networkx_labels(innovation_network, pos, labels, font_size=9, font_color='black')

# Añadir leyenda de roles (para entender los colores)
legend_handles = []
for role, color in node_colors.items():
    legend_handles.append(plt.Line2D([0], [0], marker='o', color='w', label=role,
                                     markerfacecolor=color, markersize=10))
plt.legend(handles=legend_handles, title="Roles", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.title("Red de Innovación con Tamaño de Nodo por Prestigio (20 Nodos con Roles)", size=18)
plt.axis('off') # Ocultar los ejes
plt.tight_layout() # Ajusta el layout para que la leyenda no se solape
plt.show()
````

## Modelos de Red Básicos y sus Implicaciones para la Innovación

### 1. Modelo de Erdös-Rényi (E-R): Redes Aleatorias

El modelo E-R es el más simple de entender. Imagina un grupo de nodos y una probabilidad $p$ de que exista un enlace entre cualquier par de ellos. Cada conexión es independiente de las demás.

**Conceptos Básicos:**
* **Formación Aleatoria:** Los enlaces se forman de manera completamente aleatoria e independiente.
* **Grado Homogéneo:** La mayoría de los nodos tienen un número similar de conexiones (grado), lo que significa que no hay nodos "súper conectados" o "hubs" muy pronunciados.
* **Distancia Corta:** A medida que la red crece, la distancia promedio entre nodos tiende a ser corta.

**Ventajas Teóricas para la Innovación:**
* **Difusión Rápida a Corta Distancia:** Si la red es lo suficientemente densa, las ideas pueden viajar rápidamente entre cualquier par de nodos debido a las distancias cortas. Esto podría favorecer la **difusión global de innovaciones no complejas**.
* **Resiliencia a Fallos Dirigidos:** Como no hay hubs obvios, eliminar un nodo aleatorio tiene un impacto limitado en la conectividad general. Esto podría implicar que la red es **robusta frente a la pérdida de un actor específico**, ya que la conectividad no depende de unos pocos.

**Desventajas Teóricas para la Innovación:**
* **Falta de Estructura de Comunidad:** La aleatoriedad no favorece la formación de comunidades o clústeres densos, que son cruciales para el **intercambio de conocimiento tácito** y la **construcción de confianza**. La innovación compleja y el aprendizaje suelen requerir interacciones frecuentes y cercanas.
* **Ineficiencia en la Búsqueda:** Sin nodos centrales, encontrar información específica o el conocimiento adecuado puede ser menos eficiente, ya que no hay "expertos" o "gatekeepers" claros que faciliten el flujo.
* **Poca Resiliencia a Fallos Aleatorios Masivos:** Si se produce un fallo aleatorio a gran escala (muchos nodos se desconectan al azar), la red puede fragmentarse rápidamente.

````python
num_nodes = 100
avg_degree = 4
p_er = avg_degree / (num_nodes - 1)
G_er = nx.erdos_renyi_graph(num_nodes, p_er, seed=42) # seed para reproducibilidad
analizar_red(G_er, "Red Aleatoria (Erdos-Renyi)")
````

### 2. Modelo de Barabási-Albert (BA): Redes de Escala Libre

El modelo BA introduce el concepto de "preferential attachment" (adjunción preferencial) o "los ricos se hacen más ricos". Los nuevos nodos prefieren conectarse a nodos que ya tienen muchas conexiones.

**Conceptos Básicos:**
* **Preferencia de Conexión:** Los nodos más conectados atraen más conexiones.
* **Distribución de Grado de Ley de Potencia:** Se forman unos pocos nodos con un número extremadamente alto de conexiones (hubs) y muchos nodos con pocas conexiones.
* **Redes con "Hubs":** La presencia de estos hubs es la característica distintiva.

**Ventajas Teóricas para la Innovación:**
* **Eficiencia de Difusión:** Los hubs actúan como **supercarreteras para el flujo de información y conocimiento**. Una idea innovadora en un hub puede propagarse rápidamente a una vasta porción de la red. Esto es ideal para la **difusión rápida y de amplio alcance**.
* **Búsqueda Eficiente:** Es más fácil encontrar información o el actor adecuado, ya que los hubs sirven como **puntos de acceso centralizados** para diversos dominios de conocimiento.
* **Robustez frente a Fallos Aleatorios:** Si un nodo aleatorio falla, es probable que no sea un hub, por lo que el impacto en la conectividad general es mínimo. La red es **altamente resiliente a fallos aleatorios**.

**Desventajas Teóricas para la Innovación:**
* **Vulnerabilidad a Fallos Dirigidos:** Si un hub es el que falla (o se convierte en un cuello de botella, por ejemplo, por sobrecarga), el **impacto es catastrófico** para la conectividad y el flujo de la red. En innovación, esto significa que la pérdida de un "super-inventor" o un "financista clave" puede detener el progreso en varias áreas.
* **Riesgo de Homogeneidad:** Si la información fluye predominantemente a través de los mismos hubs, podría llevar a una **menor diversidad de ideas** o a un pensamiento de grupo, limitando la innovación disruptiva que a menudo surge de la periferia.
* **Dependencia Excesiva:** La red depende demasiado de la salud y el rendimiento de unos pocos nodos clave.


````python
m_ba = int(avg_degree / 2) if avg_degree >= 2 else 1
G_ba = nx.barabasi_albert_graph(num_nodes, m_ba, seed=42)
analizar_red(G_ba, "Red Libre de Escala (Barabási-Albert)")
````

### 3. Modelo de Watts-Strogatz (WS): Redes de Mundo Pequeño

El modelo WS busca un equilibrio entre el orden y el desorden. Combina la **alta agrupación** (similar a las redes regulares) con **caminos cortos** (similar a las redes aleatorias). Se construye a partir de una red regular con alta conectividad local, y luego se "recablean" aleatoriamente una pequeña fracción de enlaces.

**Conceptos Básicos:**
* **Alta Agrupación (Clustering):** Los nodos tienden a formar clústeres densos o comunidades (los "amigos de mis amigos son mis amigos").
* **Caminos Cortos:** A pesar de la alta agrupación, la distancia promedio entre nodos es sorprendentemente corta debido a unos pocos "atajos" aleatorios (los enlaces recableados).
* **Equilibrio Local-Global:** Captura la combinación de conectividad local fuerte y conectividad global eficiente.

**Ventajas Teóricas para la Innovación:**
* **Ideal para Conocimiento Tácito y Confianza:** La alta agrupación facilita la **interacción repetida y el desarrollo de confianza** dentro de las comunidades. Esto es crucial para el **intercambio de conocimiento tácito** y la colaboración en proyectos complejos de innovación.
* **Propagación Eficiente de la Novedad:** Los caminos cortos garantizan que, a pesar de la formación de grupos, las **nuevas ideas pueden difundirse rápidamente** a través de toda la red, saltando entre comunidades a través de los "atajos".
* **Robustez Equilibrada:** Ofrece un buen balance entre la robustez a fallos aleatorios (debido a la alta redundancia local) y la vulnerabilidad a ataques dirigidos, sin la dependencia extrema de los hubs del modelo BA.

**Desventajas Teóricas para la Innovación:**
* **Menos Eficiente que BA para Difusión Masiva:** Aunque los caminos son cortos, puede que no sea tan eficiente como una red de escala libre para la **difusión explosiva** desde un único punto, ya que carece de los hubs superconectados.
* **Identificación de "Puentes" Crítica:** La eficiencia depende en gran medida de esos pocos enlaces recableados (puentes). Identificar y mantener estos **"puentes" entre comunidades** es esencial; su pérdida puede afectar la difusión global.

````python
# --- PASO 1: CALCULAR E IDENTIFICAR NODOS CRÍTICOS ---
# ¡EJECUTA ESTE BLOQUE DESPUÉS DEL PASO 0!

print("\n--- 1. Calculando Centralidades para Identificar Nodos Críticos ---")

# 1. Calcular Centralidad de Grado (Degree Centrality)
# Mide el número de conexiones directas de un nodo.
degree_centrality = nx.degree_centrality(innovation_network)

# 2. Calcular Centralidad de Intermediación (Betweenness Centrality)
# Mide cuántas veces un nodo actúa como puente en el camino más corto entre otros nodos.
betweenness_centrality = nx.betweenness_centrality(innovation_network)

# 3. Calcular Centralidad de Cercanía (Closeness Centrality)
# Mide qué tan cerca está un nodo de todos los demás en la red.
closeness_centrality = nx.closeness_centrality(innovation_network)

# 4. Calcular Centralidad de Vector Propio (Eigenvector Centrality)
# Mide la influencia de un nodo, basada en la influencia de sus vecinos.
# Puede fallar si la red no está fuertemente conectada; usamos try-except.
try:
    eigenvector_centrality = nx.eigenvector_centrality(innovation_network, max_iter=1000)
except nx.PowerIterationConvergenceFailure:
    print("Advertencia: No se pudo calcular Eigenvector Centrality (la red puede no estar fuertemente conectada o iteraciones insuficientes). Estableciendo a 0.0.")
    eigenvector_centrality = {node: 0.0 for node in innovation_network.nodes()}


# Recopilar todos los datos en un DataFrame para facilitar el análisis y la visualización
nodes_criticality_df = []
for node in innovation_network.nodes():
    role = innovation_network.nodes[node].get('role', 'Desconocido')
    nodes_criticality_df.append({
        'Nodo': node,
        'Rol': role,
        'Grado': degree_centrality.get(node, 0),
        'Intermediación': betweenness_centrality.get(node, 0),
        'Cercanía': closeness_centrality.get(node, 0),
        'Vector Propio': eigenvector_centrality.get(node, 0)
    })

critical_nodes_table = pd.DataFrame(nodes_criticality_df)

print("\n--- Tabla de Centralidades de Nodos (Ordenada por Intermediación) ---")
# Los nodos con mayor intermediación son tus "puentes" clave.
print(critical_nodes_table.sort_values(by='Intermediación', ascending=False).to_string())

print("\n--- Tabla de Centralidades de Nodos (Ordenada por Grado) ---")
# Los nodos con mayor grado son tus "hubs" o "personas más conectadas".
print(critical_nodes_table.sort_values(by='Grado', ascending=False).to_string())


print("\n--- Interpretación Inicial para 'A Quién Cuidar' ---")
print("1. **Mayor Intermediación:** Estos nodos son los 'puentes' críticos. Si se les elimina, se rompen los caminos entre diferentes partes de la red. Son los objetivos principales para 'romper' la red.")
print("   -> **Acción como Gestor:** Fortalecer las conexiones *alrededor* de estos nodos (crear redundancia), o asegurar que su rol sea replicable/distribuido.")
print("2. **Mayor Grado:** Estos nodos son los 'hubs' de la red. Son grandes difusores de información. Si se les elimina, se reduce la conectividad general y la rapidez de difusión.")
print("   -> **Acción como Gestor:** Asegurarse de que el conocimiento que poseen no sea un cuello de botella, fomentar la comunicación directa entre sus contactos clave.")

print("\n¡Ahora pasaremos a la simulación de ataque para ver el impacto real de la eliminación de estos nodos!")
````

````python
# --- PASO 2: SIMULAR ATAQUES Y EVALUAR LA RESILIENCIA ---
# ¡EJECUTA ESTE BLOQUE DESPUÉS DEL PASO 1!

import random # Necesario para el ataque aleatorio

def simulate_network_attack(network, num_nodes_to_remove, attack_type, show_plot=True):
    """
    Simula un 'ataque' a la red eliminando nodos y mide el impacto.

    Args:
        network (nx.Graph): La red original de NetworkX.
        num_nodes_to_remove (int): El número de nodos a eliminar.
        attack_type (str): 'targeted_betweenness' para eliminar los nodos con mayor intermediación,
                           o 'random' para eliminar nodos al azar.
        show_plot (bool): Si es True, muestra el gráfico de la red después del ataque.

    Returns:
        dict: Un diccionario con métricas del impacto del ataque.
    """
    print(f"\n--- Simulación de Ataque a la Red ({attack_type.replace('_', ' ').capitalize()}) ---")

    original_num_nodes = network.number_of_nodes()
    original_num_edges = network.number_of_edges()

    # Crear una copia de la red para no modificar la original
    attacked_network = network.copy()

    nodes_to_remove = []

    if attack_type == 'targeted_betweenness':
        # Calcular Centralidad de Intermediación en la red ORIGINAL
        betweenness_centrality = nx.betweenness_centrality(network)
        sorted_nodes = sorted(betweenness_centrality.items(), key=lambda item: item[1], reverse=True)
        nodes_to_remove = [node for node, bc_val in sorted_nodes[:num_nodes_to_remove]]
        print(f"Nodos a eliminar (mayor intermediación): {nodes_to_remove}")

    elif attack_type == 'random':
        all_nodes = list(network.nodes())
        random.seed(42) # Para reproducibilidad
        nodes_to_remove = random.sample(all_nodes, num_nodes_to_remove)
        print(f"Nodos a eliminar (aleatorios): {nodes_to_remove}")

    else:
        print("Tipo de ataque no reconocido. Use 'targeted_betweenness' o 'random'.")
        return {}

    # Eliminar los nodos de la red atacada
    attacked_network.remove_nodes_from(nodes_to_remove)

    # Medir el impacto
    impact_metrics = {
        'attack_type': attack_type,
        'nodes_removed_list': nodes_to_remove, # Para saber exactamente cuáles se quitaron
        'original_num_nodes': original_num_nodes,
        'original_num_edges': original_num_edges,
        'nodes_removed': num_nodes_to_remove,
        'nodes_remaining': attacked_network.number_of_nodes(),
        'edges_remaining': attacked_network.number_of_edges()
    }

    connected_components = list(nx.connected_components(attacked_network))
    num_components = len(connected_components)
    impact_metrics['num_connected_components'] = num_components

    if num_components > 0:
        lcc_size = len(max(connected_components, key=len))
        impact_metrics['largest_connected_component_size'] = lcc_size
    else:
        impact_metrics['largest_connected_component_size'] = 0

    if lcc_size > 1:
        lcc_subgraph = attacked_network.subgraph(max(connected_components, key=len))
        try:
            impact_metrics['average_path_length_lcc'] = nx.average_shortest_path_length(lcc_subgraph)
        except nx.NetworkXError: # Puede ocurrir si la LCC no es realmente un camino continuo
            impact_metrics['average_path_length_lcc'] = float('inf')
    else:
        impact_metrics['average_path_length_lcc'] = float('nan')


    print("\n--- Métricas de Impacto del Ataque ---")
    for key, value in impact_metrics.items():
        if key not in ['attack_type', 'nodes_removed_list']:
            if isinstance(value, float) and np.isnan(value):
                print(f"{key.replace('_', ' ').capitalize()}: No aplicable")
            elif isinstance(value, float):
                print(f"{key.replace('_', ' ').capitalize()}: {value:.2f}")
            else:
                print(f"{key.replace('_', ' ').capitalize()}: {value}")
    print(f"Nodos eliminados (específicos): {impact_metrics['nodes_removed_list']}")


    if show_plot:
        plt.figure(figsize=(10, 8))
        pos = nx.spring_layout(attacked_network, seed=50) # Mismo seed para comparar
        nx.draw_networkx_nodes(attacked_network, pos, node_size=700, node_color='skyblue', alpha=0.8)
        nx.draw_networkx_edges(attacked_network, pos, edge_color='gray', alpha=0.5)
        nx.draw_networkx_labels(attacked_network, pos, font_size=8, font_color='black')
        plt.title(f"Red después del Ataque ({attack_type.replace('_', ' ').capitalize()}) - Eliminados: {num_nodes_to_remove} nodos", size=15)
        plt.axis('off')
        plt.show()

    return impact_metrics


# --- CONFIGURACIÓN DE LAS SIMULACIONES ---
num_nodes_to_test = 2 # Prueba a eliminar 2 nodos. Puedes cambiar este número (ej., 3, 4)

print(f"\n--- Ejecutando Simulaciones de Ataque (eliminando {num_nodes_to_test} nodos) ---")

# Simulación 1: Ataque Dirigido (eliminando los nodos con mayor intermediación)
impact_targeted = simulate_network_attack(innovation_network, num_nodes_to_remove=num_nodes_to_test, attack_type='targeted_betweenness')

# Simulación 2: Ataque Aleatorio (eliminando el mismo número de nodos al azar)
impact_random = simulate_network_attack(innovation_network, num_nodes_to_remove=num_nodes_to_test, attack_type='random')

print("\n--- ¡Comparación Final de Impactos de Ataque! ---")
print("--------------------------------------------------")
print(f"Red Original: {impact_targeted['original_num_nodes']} nodos, {impact_targeted['original_num_edges']} aristas.")
print(f"Nodos eliminados en cada simulación: {num_nodes_to_test}\n")

print("Métrica                | Ataque Dirigido       | Ataque Aleatorio")
print("-----------------------|-----------------------|-------------------")
print(f"Nodos Restantes        | {impact_targeted['nodes_remaining']:<21} | {impact_random['nodes_remaining']}")
print(f"Aristas Restantes      | {impact_targeted['edges_remaining']:<21} | {impact_random['edges_remaining']}")
print(f"Componentes Conectados | {impact_targeted['num_connected_components']:<21} | {impact_random['num_connected_components']}")
print(f"LCC Tamaño             | {impact_targeted['largest_connected_component_size']:<21} | {impact_random['largest_connected_component_size']}")
print(f"Long. Media Caminos LCC| {impact_targeted['average_path_length_lcc'] if not np.isnan(impact_targeted['average_path_length_lcc']) else 'N/A':<21.2f} | {impact_random['average_path_length_lcc'] if not np.isnan(impact_random['average_path_length_lcc']) else 'N/A':.2f}")
print("--------------------------------------------------")

print("\n--- Conclusiones para la Resiliencia ---")
print("Compara las columnas 'Ataque Dirigido' y 'Ataque Aleatorio'.")
print("- Si el número de 'Componentes Conectados' aumenta mucho y el 'LCC Tamaño' disminuye drásticamente en el 'Ataque Dirigido' comparado con el 'Aleatorio', tu red es **vulnerable a ataques específicos** (depende de nodos clave).")
print("- Si las diferencias son pequeñas, tu red es más **resiliente** incluso ante la pérdida de nodos importantes.")
print("- Revisa los 'Nodos eliminados (específicos)' del ataque dirigido para saber **a quiénes debes cuidar** y fortalecer sus conexiones o roles.")
````
